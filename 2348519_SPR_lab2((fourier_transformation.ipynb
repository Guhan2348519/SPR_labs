Discrete-Time Fourier Transform (DTFT) and plot the con tinuous frequency spectrum. Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum.


The discrete Fourier transform (DFT) converts a finite sequence of equally-spaced samples of a function into a same-length sequence of equally-spaced samples of the discrete-time Fourier transform (DTFT), which is a complex-valued function of frequency. The interval at which the DTFT is sampled is the reciprocal of the duration of the input sequence.

```python
# Compute the DTFT
def compute_dtft(x, s_rate):
    X = np.fft.fftshift(np.fft.fft(x, 1024))  # Zero-padding for better resolution
    f = np.linspace(-s_rate/2, s_rate/2, len(X))  # Frequency vector
    return f, X

f_dtft, X_dtft = compute_dtft(sine, s_rate)

# Plot the DTFT
plt.figure()
plt.plot(f_dtft, np.abs(X_dtft))
plt.title('DTFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.show()

# Compute the DFT
X_dft = np.fft.fft(sine)
frequencies_dft = np.fft.fftfreq(len(sine), 1/s_rate)

# Plot the DFT
plt.figure()
plt.stem(frequencies_dft, np.abs(X_dft))
plt.title('DFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```

```python
# Parameters
N = 64  # Number of samples
n = np.arange(N)
f1 = 4 / N  # Frequency 1
f2 = 5 / N  # Frequency 2

# Composite signal
x = np.cos(2 * np.pi * f1 * n) + 0.5 * np.cos(2 * np.pi * f2 * n)

# Plot the composite signal
plt.plot(n, x)
plt.title('Composite Signal')
plt.xlabel('Sample index')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Compute DTFT
def dtft(x, N):
    X = np.fft.fft(x, 1024)  # Use zero-padding for better frequency resolution
    X = np.fft.fftshift(X)  # Shift zero frequency component to the center
    f = np.linspace(-0.5, 0.5, len(X))  # Normalized frequency range
    return f, X

# Get DTFT
f, X = dtft(x, N)

# Plot the continuous frequency spectrum
plt.plot(f, np.abs(X))
plt.title('Continuous Frequency Spectrum')
plt.xlabel('Normalized Frequency')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
# Perform DFT
X = np.fft.fft(x)

# Frequency axis
k = np.arange(N)
f = k / N

# Plot the magnitude spectrum
plt.stem(f, np.abs(X))
plt.title('Magnitude Spectrum')
plt.xlabel('Normalized Frequency')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Define parameters
t = np.arange(0, 1, 0.001)
a = 60
b = 2
f = 10
# Generate the signal
y = a * np.exp(-b * t) * np.sin(2 * np.pi * f * t)
# Plot the signal
plt.plot(t, y)
plt.title('Exponentially Decaying Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python

# Compute DTFT
omega = np.linspace(-np.pi, np.pi, 1000)
X_dtft = np.array([np.sum(y * np.exp(-1j * w * t)) for w in omega])

# Plot DTFT
plt.figure()
plt.plot(omega, np.abs(X_dtft))
plt.title('DTFT of Exponentially Decaying Signal')
plt.xlabel('Frequency (rad/sample)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
# Compute DFT
N = len(y)
X_dft = np.fft.fft(y)

# Plot DFT
plt.figure()
plt.stem(np.arange(N), np.abs(X_dft))
plt.title('DFT of Exponentially Decaying Signal')
plt.xlabel('Frequency Bin')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rectangular_signal(t, tau):
    return np.where(np.abs(t) <= tau / 2, 1, 0)
# Parameters
tau = 1  # Width of the rectangular pulse
t = np.linspace(-2, 2, 1000)  # Time vector

# Generate rectangular signal
rect_signal = rectangular_signal(t, tau)

# Plot the signal
plt.plot(t, rect_signal)
plt.title('Rectangular Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
